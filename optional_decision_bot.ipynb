# Optional: Decision-making Bot

Analyze pros/cons for decisions like 'Should I buy this product?'

It helps us take a decision like buying something or not, by listing pros and cons, and then giving a final recommendation

```python
# pip install -U langchain langchain-core langchain-groq python-dotenv

import os
from dotenv import load_dotenv
# Load .env
load_dotenv()

# Read API key
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
# Install if needed
# !pip install langgraph langchain-groq

from langgraph.graph import StateGraph, END
from langchain_groq import ChatGroq
import os
from typing import TypedDict, List



# Define state properly
class DecisionState(TypedDict):
    decision: str
    pros: List[str]
    cons: List[str]
    analysis: str
    recommendation: str

# Nodes
def gather_pros(state: DecisionState):
    state["pros"] = ["Benefit 1", "Benefit 2", "Benefit 3"]
    return state

def gather_cons(state: DecisionState):
    state["cons"] = ["Drawback 1", "Drawback 2"]
    return state

def analyze_decision(state: DecisionState):
    pros_count = len(state["pros"])
    cons_count = len(state["cons"])

    if pros_count > cons_count:
        state["recommendation"] = "Yes"
    else:
        state["recommendation"] = "No"

    state["analysis"] = (
        f"Pros: {pros_count}, "
        f"Cons: {cons_count}. "
        f"Recommendation: {state['recommendation']}"
    )

    return state  


# Build graph
graph = StateGraph(DecisionState)

graph.add_node("pros", gather_pros)
graph.add_node("cons", gather_cons)
graph.add_node("analyze", analyze_decision)

graph.set_entry_point("pros")

graph.add_edge("pros", "cons")
graph.add_edge("cons", "analyze")
graph.add_edge("analyze", END)

# Compile
app = graph.compile()

# Run
result = app.invoke({
    "decision": "Buy laptop",
    "pros": [],
    "cons": [],
    "analysis": "",
    "recommendation": ""
})

print(result["analysis"])
```
