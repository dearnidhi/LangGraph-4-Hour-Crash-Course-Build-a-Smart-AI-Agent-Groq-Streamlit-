## LangChain Recap
LangChain is a framework for building applications with LLMs.

### Key Concepts:
- **Agents**: Agents are AI systems that can decide which tools to use.   
    They can:
    - understand the task
    - choose a tool
    - use the tool
    - generate a response

- **Chains**: Sequences of calls to LLMs or other utilities.    
    A chain is a sequence of steps.   
    Example:- Input → Prompt → AI Model → Output        
    Everything happens step by step in a fixed flow.


- **Tools**: Functions that agents can call to perform actions
    Tools are functions the AI can use to perform actions.

    Examples:
    - Search the internet
    - Calculate math
    - Read files
    - Get current time
    
### Simple Understanding

- Chains = Fixed workflow
- Tools = Actions/functions
- Agents = Decision makers


### Example: Simple Chain

`````python
# pip install -U langchain langchain-core langchain-groq python-dotenv

import os
from dotenv import load_dotenv

from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq

# Load .env
load_dotenv()

# Read API key
groq_api_key = os.getenv("GROQ_API_KEY")

# Initialize Groq model
llm = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama-3.1-8b-instant",
    temperature=0
)

# Prompt
prompt = PromptTemplate.from_template(
    "Translate the following text to French:\n\n{text}"
)

# Create chain
chain = prompt | llm

# Invoke
result = chain.invoke({
    "text": "Hello"
})

# Output
print(result.content)
`````

### Agents Example

`````python
# Example agent (simplified)
def simple_agent(query):
    if  "time" in query.lower():
        return "Current time tool:" + str(__import__('datetime').datetime.now())
    else:
        return "LLM response: I dont know"
print(simple_agent("What is the current time?"))  

print(simple_agent("hello?"))
`````

## What is LangGraph?

LangGraph is a library used to build smart AI workflows using LLMs.

You can think of LangGraph like a flowchart for AI applications.

Instead of running only one fixed sequence,
LangGraph allows the AI application to:
- make decisions
- use memory
- call tools
- retry steps
- work with multiple agents

This makes it useful for building advanced AI systems.


### Key Features

1. Stateful Workflows

LangGraph can remember information between steps.

Example:
A chatbot remembers previous messages.

This memory is called "state".

---

2. Branching Logic

LangGraph can make decisions.

Example:
- If user asks about weather → use weather tool
- Else → use LLM response

So the workflow can change dynamically.

---

3. Multi-actor Systems

Multiple agents or components can work together.

Example:
- One agent does research
- Another agent summarizes results
- Another agent checks quality

---

4. Persistence

LangGraph can save and load state.

This means:
- conversations can continue later
- workflows can resume after stopping


### Common Use Cases

- AI Chatbots
- Research Agents
- Customer Support Bots
- Decision-making Systems
- AI Assistants

## Why LangGraph?

Simple chains are good for basic tasks.

But real AI applications are more complex.

Sometimes AI needs:
- memory
- decision making
- retries
- multiple steps
- human approval

LangGraph helps manage these complex workflows easily.

### Why Developers Use LangGraph

- Better memory management
- Flexible workflows
- Error handling and retries
- Multi-agent systems
- Human-in-the-loop support

## Graph vs Simple Chain

### Simple Chain

A simple chain follows one fixed path.

Example:

A → B → C

Everything happens step by step.

There are:
- no decisions
- no loops
- no branching

Good for simple tasks.

---

### Graph

A graph is more flexible.

It uses:
- Nodes → steps/functions
- Edges → connections between steps
- State → shared memory

A graph can:
- make decisions
- branch into different paths
- repeat steps using loops
- store memory
- run multiple components

Example:
```
A → B
↓   ↑
C → D
```
This makes LangGraph powerful for advanced AI workflows.

## Workflow Example
```
Start
   ↓
Process Input
   ↓
Decide Action
   ↓
Tool or LLM
   ↓
Update Memory
   ↓
End
````

If the response is unclear,
the workflow can loop back and try again.

This is something normal chains cannot do easily.
