# Optional: Multi-step Research Agent

Build an agent that breaks queries into sub-questions, searches, summarizes.

This agent breaks a big question into smaller sub-questions,  
 "searches" each one (simulated here), and then summarizes everything.




```python
# pip install -U langchain langchain-core langchain-groq python-dotenv

import os
from dotenv import load_dotenv
# Load .env
load_dotenv()

# Read API key
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python

from langgraph.graph import StateGraph, END

# -----------------------------
# State
# -----------------------------
class ResearchState(dict):
    pass

# -----------------------------
# Nodes
# -----------------------------
def break_down_query(state):
    state["sub_questions"] = [
        f"What is {state['query']}?",
        f"History of {state['query']}",
        f"Current trends in {state['query']}"
    ]
    return state

def search1(state):
    question = state["sub_questions"][0]
    answer = f"Simulated answer for: {question}"

    state["answers"].append(answer)

    return state


def search2(state):
    question = state["sub_questions"][1]
    answer = f"Simulated answer for: {question}"

    state["answers"].append(answer)

    return state

def search3(state):
    question = state["sub_questions"][2]
    answer = f"Simulated answer for: {question}"

    state["answers"].append(answer)

    return state

def summarize_answers(state):
    summary = "Summary:\n\n" + "\n".join(state["answers"])

    state["final_answer"] = summary

    return state



# -----------------------------
# Graph
# -----------------------------
graph = StateGraph(dict)

graph.add_node("break_down", break_down_query)
graph.add_node("search1", search1)
graph.add_node("search2", search2)
graph.add_node("search3", search3)
graph.add_node("summarize", summarize_answers)

# Entry
graph.set_entry_point("break_down")

# Flow
graph.add_edge("break_down", "search1")
graph.add_edge("search1", "search2")
graph.add_edge("search2", "search3")
graph.add_edge("search3", "summarize")

graph.add_edge("summarize", END)

# Compile
app = graph.compile()

# Run
result = app.invoke({
    "query": "AI",
    "sub_questions": [],
    "answers": [],
    "final_answer": ""
})

print(result["final_answer"])    
```
